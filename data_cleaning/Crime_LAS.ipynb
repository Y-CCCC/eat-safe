```python
# csv is a package for file reading/IO
import csv

# pandas is the data analysis library we'll use in this course
import pandas as pd
```

```python
crimelas_df = pd.read_csv('Las_Vegas_Metropolitan_Police_calls_for_service.csv', header = None, names = ['Event Number', 'Event Date', 'Type', 'Type Description',\
                                                                     'General Location', 'Beat', 'Disposition','Map X', 'Map Y', \
                                                                      'Location 1', 'Ward', 'Zip Codes'],dtype="object")
```

```python
crimelas_df = crimelas_df.drop(['Type', 'Type Description', 'Beat', 'Disposition','Map X', 'Map Y', \
                                            'Ward', 'Zip Codes'], axis=1)
```
